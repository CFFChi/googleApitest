```python
i = 0
failLink =[]
siteDict = {}
TimeoutError = []
for t in range(len(RList)):
#     if t ==0:
#         t = 457
    ele = RList[t]
# for ele in RList:
    ele = ele.replace("\n","")
    try:
#         print(ele)
        req = urllib.request.Request(
        ele,
        headers = {
                "User-Agent":random.choice(user_agent_list)
            }

        )
        response = urllib.request.urlopen(req)
        if response.info().get("Content-Encoding") == "gzip":
            fzip = BytesIO(response.read())
            page = gzip.GzipFile(fileobj = fzip).read()
        else:
            page = urllib.request.urlopen(req).read().decode("utf-8","replace")
        soup = bs(page,"html.parser")
        menulink = []
        sitemap = []
        tempDict = {}

        for link in soup.find_all("a"):
            if "menu" in str(link.get("href")).lower() or "menu" in link.get_text().lower():
# #                 print(link.get("href"))
#                 print("-"*30)
#                 print(urljoin(ele,link.get("href")))
                menulink.append(urljoin(ele,link.get("href")))
            if "site" in str(link.get("href")).lower() and "map" in str(link.get("href")).lower():
                sitemap.append(urljoin(ele,link.get("href")))
                
        menulink = list(set(menulink))
        if "#" in menulink:
            menulink.remove("#")
#         print(set(menulink))
#         print(sitemap)
     
        if len(menulink) is 0 and len(sitemap) is 0:
            failLink.append(ele)
#             print("add here fail 1")
       
        if ele not in failLink:
            if "menulink" not in tempDict.keys():
                tempDict["menulink"] = []
            tempDict["menulink"].extend(menulink)

            if "sitemap" not in tempDict.keys():
                tempDict["sitemap"] = []
            tempDict["sitemap"].extend(sitemap)

            if ele not in siteDict.keys():
                siteDict[ele] = []
            siteDict[ele].append(tempDict)
#             print("add siteDict 1")
#         print(i)
#         print("*"*30)
        i +=1

            
            
    except URLError:
        failLink.append(ele)
#         print("add here fail 2")
        i +=1
#         print("*"*30)
    except IncompleteRead as e:     
        pageS = e.partial
        soupS = bs(pageS,"html.parser")
        menulink = []
        sitemap = []
        tempDict = {}
        
        for link in soupS.find_all("a"):
            if(link.get("href") is not None):
                if "menu" in str(link.get("href")).lower() or "menu" in link.get_text().lower():
#                     print(urljoin(ele,link.get("href")))    
                    menulink.append(urljoin(ele,link.get("href")))
                if "site" in str(link.get("href")).lower() and "map" in str(link.get("href")).lower():
                    sitemap.append(urljoin(ele,link.get("href")))

        menulink = list(set(menulink))
        if "#" in menulink:
            menulink.remove("#")


        if len(menulink) is 0 and len(sitemap) is 0:
            failLink.append(ele)
#             print("add here fail 3")

        if ele not in failLink:
            if "menulink" not in tempDict.keys():
                tempDict["menulink"] = []
            tempDict["menulink"].extend(menulink)

            if "sitemap" not in tempDict.keys():
                tempDict["sitemap"] = []
            tempDict["sitemap"].extend(sitemap)

            if ele not in siteDict.keys():
                siteDict[ele] = []
            siteDict[ele].append(tempDict)
#             print("add siteDict 2")
        print(i)
#         print("*"*30)
        i +=1

    except:
        failLink.append(ele)
                        
print(len(failLink))       
print(len(siteDict))
```
