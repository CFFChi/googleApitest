```python
# get all text in leaves of this node
def print_Text(node):
    stack = []
    stack.append(node)
    Text = []
    while len(stack) is not 0:
        Parent = stack.pop()
        Children = Parent.findAll(recursive = False)
        
        if len(Children) is not 0:
            for child in Children:
                stack.append(child)
            textTemp = Parent.findAll(text= lambda text:isinstance(text, Comment), recursive = False)
            [comment.extract() for comment in textTemp]
            
            textTemp = Parent.findAll(text = True, recursive = False)
            if len(textTemp) is not 0:
                for ele in textTemp:
#                     print("parent text:{}".format(ele))
                    if ele !="\n" and ele!=" ":
                        ele = ele.replace("\t","").replace("\n","").replace("\r","").replace("  "," ").replace("\xa0"," ")
                        Text.append(ele)
        else:
            if "script" not in Parent.name and len(Parent.get_text()) is not 0:

                textTemp2 = Parent.get_text().replace("\t","")
                textTemp2 = textTemp2.replace("\n","").replace("\r","").replace("  "," ").replace("\xa0"," ")
                Text.append(textTemp2)
#                 print(Parent.get_text())
    return Text
def print_Text2(node):
    Text = []
    textTemp = node.findAll(text= lambda text:isinstance(text, Comment))
    [comment.extract() for comment in textTemp]

    textTemp = node.findAll(text = True)
    if len(textTemp) is not 0:
        for ele in textTemp:
#                     print("parent text:{}".format(ele))
            if ele !="\n" and ele!=" ":
                ele = ele.replace("\t","").replace("\n","").replace("\r","").replace("  ","").replace("\xa0"," ")
                if ele != " " and ele !="":
                    Text.append(ele)
                
    return Text
```

```python
# weburl = "https://www.bostonmarket.com/menu/desserts/carrot-cake/"
weburl = "http://thorntonsfenwaygrille.com/menus/"
# weburl1 = "https://www.HimalayanKitchenMA.com/#menu"
# weburl1 = "http://thebrahmin.com/menus/"
# weburl = "http://paulisnorthend.com/catering-menu/"
def getMenu(weburl):
    Menu = []
    
    req = urllib.request.Request(
            weburl,
            headers = {
                    "User-Agent":random.choice(user_agent_list)
                }

            )

    response = urllib.request.urlopen(req)
    if response.info().get("Content-Encoding") == "gzip":
        fzip = BytesIO(response.read())
        page = gzip.GzipFile(fileobj = fzip).read()
    else:
        page = urllib.request.urlopen(req).read().decode("utf-8","replace")
    soup = bs(page,"html.parser")
    nowTag = soup.find("body")
    if nowTag is None:
        return
    Thro = 5
    
#     print(len(nowTag.findAll(recursive=False)))
#     print(nowTag.name)
    # for ele in nowTag.findAll(recursive=False):\
    stack = []
    stack.append(nowTag)

    MenuNode = []

    while len(stack) is not 0:
        Parent = stack.pop()
        Children = Parent.findAll(recursive = False)
        # into stack 
        if len(Children) is not 0:
            Tag = []
            for child in Children:
                stack.append(child)
    #             print("attrs:{}".format(child.attrs))
    #             print("name:{}".format(child.name))
                Tag.append(child.name)
            TagDict = Counter(Tag).most_common(1)
    #         print(TagDict)
    #         print("*"*30)

            if TagDict[0][1]>= Thro and "script" not in TagDict[0][0].lower():
#                 print(TagDict[0][0]) 
                MenuNode.append(Parent)

#     print("*"*30)
    for node in MenuNode:
        FS = []
        MN = []
#         print("nodeNmae:{}".format(node.name))
        if "body" in node.name:
            continue
        
        PsibList=node.findPreviousSiblings()
        textPreSib = []
        for Psib in PsibList :
#             print(Psib.name)
#             print(Psib.get_text())
#             print(len(Psib.get_text()))
            textPreSib.extend(print_Text2(Psib))
        if len(textPreSib) is not 0:
#             print("foodScore:{}".format(foodScore(textPreSib)))
#             print("MoneyNum:{}".format(MoneyNum(textPreSib)))
            FS.append(foodScore(textPreSib))
            MN.append(MoneyNum(textPreSib))

        
        textNode = print_Text2(node)

        if len(textNode) is not 0:
#             print("foodScore:{}".format(foodScore(textNode)))
#             print("MoneyNum:{}".format(MoneyNum(textNode)))

            FS.append(foodScore(textNode))
            MN.append(MoneyNum(textNode))

#         print(node.get_text())
#         print(len(node.get_text()))
        
    
        NsibList = node.findNextSiblings()
        textNexSib = []
        for Nsib in NsibList:
#             print(Nsib.name)
            textNexSib.extend(print_Text2(Nsib))
#         print("Next Sib:\n {}".format(textNexSib))
        if len(textNexSib) is not 0:
#             print("foodScore:{}".format(foodScore(textNexSib)))
#             print("MoneyNum:{}".format(MoneyNum(textNexSib)))

            FS.append(foodScore(textNexSib))
            MN.append(MoneyNum(textNexSib))
        menuFlag = False
        for i in range(len(FS)):
            if FS[i]>=0.15 and MN[i] > 2:
                menuFlag = True
                break
        if len(FS) is not 0:
            maxFS = max(FS)
            if maxFS>=0.2:
                if len(textPreSib) is not 0:
                    Menu.append(textPreSib)
                if len(textNode) is not 0:
                    Menu.append(textNode)
                if len(textNexSib) is not 0:
                    Menu.append(textNexSib)
            elif menuFlag:
                if len(textPreSib) is not 0:
                    Menu.append(textPreSib)
                if len(textNode) is not 0:
                    Menu.append(textNode)
                if len(textNexSib) is not 0:
                    Menu.append(textNexSib)

#         print("*"*30)
#         print("PreSib: {}".format(textPreSib))

#         print("this node:{}".format(textNode))
#         print(len(Nsib.get_text()))
    
    return Menu
    # brother = nowTag.findAll(recursive=False)[0]
    # RepeatedTag(brother, Thro)

getMenu("http://artuboston.com/menu-item-type/beacon-hill-lunch/")
```
